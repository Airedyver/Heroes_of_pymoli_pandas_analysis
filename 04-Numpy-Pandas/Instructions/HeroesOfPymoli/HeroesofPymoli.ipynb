```python
#Break into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
bins = (0,10,14,19,23,27,31,35,39,43,47)
# Create the names for the four bins
group_names = ['0-10', '11-14', '15-19', '20-23', '24-27', '28-31', '32-35', '36-39','40-43','44-47']

# Cut postTestScore and place the scores into bins
age_bins = pd.cut(purch_one_df["Age"], bins, labels=group_names)



```

```python
#add age bins as another category in the dataframe

purch_one_df["Age Group"] = age_bins
purch_one_df.head()
```

```python
# Purchase Count, get the Average Purchase Price

age_analysis = purch_one_df[["Age Group", "Price"]]
age_group = age_analysis.groupby(["Age Group"])
age_group.mean()

 
 
```
